```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
x_axis = []
y_axis = []
for i in range(158):
    x_axis.append(i)
    value_y = math.sqrt(i/(4*(math.pi**2 *(5*10**-9)*(1.6*10**-19))))
    y_axis.append(value_y)

plt.ylabel('frequency(Hz)')
plt.xlabel('voltage(V)')
plt.title('no.1e')
plt.plot(x_axis,y_axis)
```
